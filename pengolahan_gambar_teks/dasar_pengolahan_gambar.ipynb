```python
import matplotlib.pyplot as plt
import cv2
```

```python
# Image data reading
print("read images using opencv")
five = cv2.imread("gambar_pert2/5.png")
print(five.shape)
print(five.size)
plt.imshow(five)
cv2.waitKey(0)
```

```python
# image conversion to GRAY
babon = cv2.imread("gambar_pert2/babon.jpg")
babon_gray = cv2.cvtColor(babon, cv2.COLOR_BGR2GRAY)

plt.imshow(babon) 
plt.imshow(babon_gray)
```

```python
# Image cropping. hanya mengambil gambar hidung Babon
roi = babon[150:180, 70:150]
plt.imshow(roi)
```

```python
# Image resizing. Resize ke ukuran 100 x 100 pixel
resized_babon = cv2.resize(babon, (100, 100))
plt.imshow(resized_babon)
```

```python
# mendapatkan ukuran Gambar
(h, w, d) = resized_babon.shape
h,w,d
```

```python
# Image Rotation.
center = (w // 2, h // 2)
M = cv2.getRotationMatrix2D(center, -45, 1.0)
rotated_babon = cv2.warpAffine(resized_babon, M, (w, h))
plt.imshow(rotated_babon)
```

```python
# Smoothing image with Gaussian Blur dengan kernel ukuran 11 x 11
blurred_babon = cv2.GaussianBlur(resized_babon, (11, 11), 0)
plt.imshow(blurred_babon)
```
